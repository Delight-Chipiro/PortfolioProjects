# Autotrader Web Scraping 

```python
from bs4 import BeautifulSoup 
import requests
import time
import smtplib
import csv
```

```python
URL = 'https://www.autotrader.co.za/car-for-sale/toyota/hilux/2.4gd-6/27465266?vf=2&db=1&s360=0&so=1&pl=1&pq=0&pr=3&po=1'

page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')

print(soup)
```

```python
name = soup.find(class_="e-listing-title").get_text()

print(name)
```

```python
price = soup.find(class_ ="e-price").get_text()

p = price.strip()[2:]

string_price = p

#print(price)

print(string_price)
```

```python
import re

fixed = re.sub(r'\s+','',string_price)

real_price = int(fixed)

print(real_price)
```
